```python
# y=beta0 + beta1*x1 +beta2*x2 .......+betan*xn + e(error)
# y = 3 + 2x1 + 4*x2 + error
```

```python
y= 3 + 2*x1 + 4*x2 + np.random.random(50)
```

```python
x=pd.DataFrame({"Intercept":np.ones_like(50),'x1':x1,'x2':x2})
```

```python
w = np.zeros(x.shape[1])
```

```python
w
```

```python
def mypredictions(features,weights):
    mypredictions = np.dot(features,weights)
    return(mypredictions)
mypredictions(x,w)
```

```python
def myerror(target,features,weights):
    error = target-mypredictions(features,weights)
    return(error)
myerror(y,x,w)
```

```python
def mycost(target,features,weights):
    error = myerror(target,features,weights)
    cost = np.dot(error.T,error)
    return(cost)
mycost(y,x,w)   
```

```python
def gradient(target,features,weights):
    error = myerror(target,features,weights)
    gradient = -np.dot(features.T,error)
    return(gradient)
gradient(y,x,w)
```

```python
def my_lr(target,features,learning_rate):
    weights=np.zeros(features.shape[1])
    pcf = mycost(target,features,weights)
    for i in np.arange(100000):
        weights -= learning_rate*gradient(target,features,weights)
        weights[0] -=30*learning_rate*gradient(target,features,weights)[0] # if you comment this out , your coefficient estimates will not be as good as sklearn, why?
        
        ncf = mycost(target,features,weights)
        if abs(ncf-pcf) < 0.00001:
            break
        if i%100==0:
            print(i,pcf,ncf,weights)
        pcf = ncf
    return(weights)

my_lr(y,x,0.0000001)
```
